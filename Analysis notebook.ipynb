## Problem Statment

**In a competitve market like the restaurant industry, understanding the factors that influence business success is crucial for stakeholders. Utilizing the Yelp datset, this project aims to investigate the relationship between user engagement(reviews, tips, and check-ins) and business success metrices (review count, ratiings) for restaurants.**
                                                                                                                                               

## Research Objectives
1. **Quantity the correlation between user engagemnt(reviews, tips, check-ins) and review count/average star rating:** This will help us determine if restaurants with higher user engagement experience a corresponding increase in reviews and ratings.
2. **Analyze the impact of sentiment in review count and average star rating:** We will Investigate if positive sentiment in reviews and tis translate to higher star ratings and potentially influences the total number of review left.
3. **Time trends in User Engagment:** We will explore if consistent user engagement over time is a stronger indicater of long-term success compared to sporadic burts of activity

## Hypothesis Testing

* Higher levels of user engagement (more reviews, tips, and check-ins) correlate with higher review counts and ratings for restaurants.
* Positive sentiment expressed in reviews and tips contributes to higher overall ratings and reviews counts for restaurants.
* Consistent engagement over time is positively associated with sustained business success for restaurants.

```python
result = pd.read_sql_query("""
SELECT
    AVG(review_count) AS avg_review_count,
    MIN(review_count) AS min_review_count,
    MAX(review_count) AS max_review_count,
    (
        SELECT AVG(review_count)
        FROM (
            SELECT review_count
            FROM business
            WHERE business_id IN ({ids})
            ORDER BY review_count
            LIMIT 2 - (SELECT COUNT(*) FROM business WHERE business_id IN ({ids})) % 2
            OFFSET (SELECT (COUNT(*) - 1) / 2 FROM business WHERE business_id IN ({ids}))
        )
    ) AS median_review_count,

    AVG(stars) AS avg_star_rating,
    MIN(stars) AS min_star_rating,
    MAX(stars) AS max_star_rating,
    (
        SELECT AVG(stars)
        FROM (
            SELECT stars
            FROM business
            WHERE business_id IN ({ids})
            ORDER BY stars
            LIMIT 2 - (SELECT COUNT(*) FROM business WHERE business_id IN ({ids})) % 2
            OFFSET (SELECT (COUNT(*) - 1) / 2 FROM business WHERE business_id IN ({ids}))
        )
    ) AS median_star_rating

FROM business
WHERE business_id IN ({ids})
""".format(
    ids=",".join(f"'{i}'" for i in business_id['business_id'])
), conn)

# Make output readable
result.T.rename(columns={0: "Value"})
```

```python
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr 
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
review_count_df = pd.read_sql_query(f"""
SELECT 
    total.avg_rating AS rating,
    AVG(total.review_count) AS avg_review_count,
    AVG(total.checkin_count) AS avg_checkin_count,
    AVG(total.tip_count) AS avg_tip_count
FROM 
(
    SELECT 
        b.business_id,
        SUM(b.review_count) AS review_count,
        AVG(b.stars) AS avg_rating,
        SUM(LENGTH(c.date) - LENGTH(REPLACE(c.date, ',', '')) + 1) AS checkin_count,
        SUM(tip.tip_count) AS tip_count
    FROM business b
    LEFT JOIN "check" c 
        ON b.business_id = c.business_id
    LEFT JOIN (
        SELECT 
            business_id, 
            COUNT(business_id) AS tip_count 
        FROM tip 
        GROUP BY business_id
    ) tip 
        ON b.business_id = tip.business_id
    WHERE b.business_id IN {tuple(business_id['business_id'])}
    GROUP BY b.business_id
) AS total
GROUP BY total.avg_rating
""", conn)
```

```python
# fucntion to calculate the success score vased on the avg rating and total review_count 
def calculate_success_score(df):
    success_score = []
    for index, row in df.iterrows():
        score = row['avg_rating'] * np.log(row['review_count'] + 1)
        success_score.append(score)
    return success_score
```

```python
# how do the success metrics (review_count or avg_rtaing) of restaurants vary accross diffrent states and cities
city_df = pd.read_sql_query(f"""select city, state, latitude, longitude, AVG(stars) as avg_rating, SUM(review_count) as review_count,
count(*) as restaurant_count
from business
WHERE business_id IN {tuple(business_id['business_id'])}
group by state, city
order by review_count desc
limit 10""", conn)

city_df['success_score'] = calculate_success_score(city_df)
```

```python
# Are there any pattern in user engagement over time for successful business compared to less successful one?
# Are there any seasonal trends in the user engagement for restaurants?

low_rated_engagement = pd.read_sql_query(f"""
SELECT
    r.month_year,
    r.review_count,
    t.tip_count
FROM
    (
        SELECT
            strftime('%m-%Y', date) AS month_year,
            COUNT(*) AS review_count
        FROM review
        WHERE business_id IN {tuple(business_id['business_id'])}
          AND stars < 3.5
        GROUP BY month_year
    ) r
LEFT JOIN
    (
        SELECT
            strftime('%m-%Y', tip.date) AS month_year,
            COUNT(*) AS tip_count
        FROM tip
        JOIN business b
            ON tip.business_id = b.business_id
        WHERE tip.business_id IN {tuple(business_id['business_id'])}
          AND b.stars < 3.5
        GROUP BY month_year
    ) t
ON r.month_year = t.month_year
ORDER BY r.month_year
""", conn)
```

```python
high_rated_engagement = pd.read_sql_query(f"""
SELECT
    r.month_year,
    r.review_count,
    t.tip_count
FROM
    (
        SELECT
            strftime('%m-%Y', date) AS month_year,
            COUNT(*) AS review_count
        FROM review
        WHERE business_id IN {tuple(business_id['business_id'])}
          AND stars >= 3.5
        GROUP BY month_year
    ) r
LEFT JOIN
    (
        SELECT
            strftime('%m-%Y', tip.date) AS month_year,
            COUNT(*) AS tip_count
        FROM tip
        JOIN business b
            ON tip.business_id = b.business_id
        WHERE tip.business_id IN {tuple(business_id['business_id'])}
          AND b.stars >= 3.5
        GROUP BY month_year
    ) t
ON r.month_year = t.month_year
ORDER BY r.month_year
""", conn)
```

```python
# How does the sentiment of reviews and tips (useful, funny,  cool) correlate with the success metrics of restaurants?

sentiment_df = pd.read_sql_query(f""" SELECT b.business_id, AVG(b.stars) as avg_rating, SUM(b.review_count) as review_count,
SUM(s.useful_count) as useful_count,
SUM(s.funny_count) as funny_count,
SUM(s.cool_count) as cool_count
FROM
( SELECT business_id,
    SUM(useful) as useful_count,
    SUM(funny) as funny_count,
    SUM(cool) as cool_count
FROM 
    review
GROUP BY business_id) as s
JOIN business as b on b.business_id = s.business_id

WHERE b.business_id IN {tuple(business_id['business_id'])}
GROUP BY b.business_id
ORDER BY review_count""", conn)

sentiment_df = remove_outliers(sentiment_df, 'review_count')
sentiment_df = remove_outliers(sentiment_df,'useful_count')
sentiment_df = remove_outliers(sentiment_df,'funny_count')
sentiment_df = remove_outliers(sentiment_df,'cool_count')
```
